# Library Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

# Loading the Video File

```python
videoFile = 'TomAndJerry.mp4'
cap = cv2.VideoCapture(videoFile)
```

# Extracting Frames

```python
frameRate = cap.get(5)
frames = []
while(cap.isOpened()):
    frameId = cap.get(1)
    success, frame = cap.read()
    if not success:
        break
    if frameId % math.floor(frameRate) == 0:
        frames.append(frame)
cap.release()
```
